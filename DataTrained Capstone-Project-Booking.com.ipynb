```python
import requests
from bs4 import BeautifulSoup

# Base URL
base_url = "https://www.booking.com/"

# List of city names
cities = [
    "Dubai, United Arab Emirates",
    "London, United Kingdom",
    "Kuala Lumpur, Malaysia",
    "Manchester",
    "New Delhi",
    "Birmingham",
    "Berlin",
    "Sydney",
    "Melbourne",
    "Paris",
    "Tokyo",
    "Toronto",
]

# Initialize a session to handle cookies and headers
session = requests.Session()

# Fetch the main Booking.com page
url = "https://www.booking.com/index.en-gb.html?label=gen173nr-1BCAEoggI46AdIM1gEaGyIAQGYAQm4ARfIAQzYAQHoAQGIAgGoAgO4AqDQ76gGwAIB0gIkMjI5NjViYjAtYTY1OS00OTYwLThmZGMtYWE3YmIxYjgzZTRi2AIF4AIB&sid=be9e7556303d905d3347178052c1d002&keep_landing=1&sb_price_type=total"
response = session.get(url)

# Parse the HTML content of the main page
soup = BeautifulSoup(response.content, "html.parser")

# Loop through the list of cities and construct their respective URLs
city_urls = {}
for city in cities:
    # Construct the URL for each city
    city_url = base_url + f'searchresults.en-gb.html?ss={city.replace(" ", "%20")}'
    
    # Store the city URL in a dictionary
    city_urls[city] = city_url

# Print the URLs for each city
for city, city_url in city_urls.items():
    print(f"{city}: {city_url}")

# Close the session
session.close()
```
