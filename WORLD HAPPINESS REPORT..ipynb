# Removing The outliers

```python
from scipy .stats import zscore
```

```python
z=np.abs(zscore(df))
z
```

```python
threshold=3
print(np.where(z>3))
```

```python
z[27][9]
```

```python
hf=df[(z<3).all(axis=1)]
hf
```

```python
df.shape
```

```python
hf.shape
```

```python
Train test split
TO make the x_train,y_train split wwe are going to split the data
```

```python
happiness_t=pd.DataFrame(hf['Happiness Score'])
```

```python
happiness_t
```

```python
y=happiness_t
y
```

```python
happiness_data=pd.DataFrame(hf)
```

```python
x=happiness_data
x
```

```python
x.shape
```

```python
y.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.25,random_state=53)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
lm=LinearRegression()     #Now we use diffrent models for prediction.
```

```python
lm.fit(x_train,y_train)
```

```python
pred=lm.predict(x_test)
print('Predicted result price:',pred)
print('actual price',y_test)
print('error:')
print(mean_squared_error(pred,y_test))
```

```python
from sklearn.linear_model import Lasso,Ridge
```

```python
ls=Lasso(alpha=0.0001)
ls.fit(x_train,y_train)
ls.score(x_train,y_train)
```

```python
rd=Ridge(alpha=0.0001)
rd.fit(x_train,y_train)
rd.score(x_train,y_train)
```

```python
from sklearn.svm import SVR

kernellist=['linear','poly','rbf']
for i in kernellist:
    sv=SVR(kernel=i)
    sv.fit(x_train,y_train)
    print(sv.score(x_train,y_train))
```
