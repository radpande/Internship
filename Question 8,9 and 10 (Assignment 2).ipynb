```python
#Q8: Write a python program to scrape data for Top 1000 Quotes of All Time.
The above task will be done in following steps:
1. First get the webpagehttps://www.azquotes.com/
2. Click on TopQuotes
3. Than scrap a) Quote b) Author c) Type Of Quotes
```

```python
import selenium
import pandas as pd
from selenium import webdriver
import warnings
warnings.filterwarnings('ignore')
from selenium.webdriver.common.by import By
import time
```

```python
driver = webdriver.Chrome(r"D:\Downloads\chromedriver_win32\chromedriver.exe")
```

```python
driver.get("https://www.azquotes.com/")
```

```python
driver.maximize_window()
```

```python
submit=driver.find_element(By.XPATH,'/html/body/div[1]/div[1]/div[1]/div/div[3]/ul/li[5]/a')
submit.click()
```

```python
quote=[]
author=[]
quote_type=[]
```

```python
quote_tags=driver.find_elements(By.XPATH,'//a[@class="title"]')
for i in quote_tags[0:100]:
    Qquotes=i.text
    quote.append(Qquotes)
```

```python
author_tags=driver.find_elements(By.XPATH,'//div[@class="author"]')
for i in author_tags[0:100]:
    authors=i.text
    author.append(authors)
```

```python
quote_tags=driver.find_elements(By.XPATH,'//div[@class="tags"]')
for i in quote_tags[0:100]:
    Quotes=i.text
    quote_type.append(Quotes)
```

```python
print(len(quote),len(author),len(quote_type))
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'Quotes':quote,'Authors':author,'Type':quote_type})
df
```

```python
driver.close()
```
